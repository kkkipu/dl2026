# Labwork 1 - Gradient Descent

```python
def f(x):
    return x**2

def df(x):
    return 2*x

def gradient_descent(start, learning_rate):
    i = 0
    x = start
    while f(x) >= 0.001:
        i += 1
        x = x - learning_rate * df(x)
        print(f"Iteration {i}: x = {x:.4f}, f(x) = {f(x):.4f}")

gradient_descent(10, 0.1)

# def gradient_descent(start, learning_rate, threshold):
#     i = 0
#     x = start
#     while True:
#         i += 1
#         f0 = f(x)
#         x = x - learning_rate * df(x)
#         f1 = f(x)
#         print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
#         if abs(f1 - f0) <= threshold:
#             break
# gradient_descent(10, 0.1, 1)
```
