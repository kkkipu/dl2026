```python
import math

def read_txt(file_path):
    with open(file_path, "r") as f:
        lines = f.readlines()
    num_layers = int(lines[0].strip())
    layer_sizes = [int(line.strip()) for line in lines[1:num_layers+1]]
    return num_layers, layer_sizes

num_layers, layer_sizes = read_txt("neural_network.txt")
print(num_layers, layer_sizes)
# 3 layers (input, hidden, output)
# 2 input layer => 2 neurons
# 2 hidden layer => 2 neurons
# 2 output layer => 1 neurons

# Activation function
def sigmoid(z):
    # return 1 / (1 + math.exp(-z))

    # With threshold (step) activation
    if z >= 0:
        return 1
    else:
        return 0
```

```python
class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
    
    def feedforward(self, inputs):  # forward = feedforward
        # z = w1*x1 + w2*x2 + 1*w0
        # w0 = bias/free coefficient
        z = 0
        for i in range(len(inputs)):
            z += inputs[i] * self.weights[i]
        z += self.bias

        # Output (activation) of neuron i in layer l
        a = sigmoid(z)
        return a


class Layer:                        # Layer has many neurons
    def __init__(self, neurons):
        self.neurons = neurons
    
    def feedforward(self, inputs):  # Send list of outputs, one from each neuron
        outputs = []

        for neuron in self.neurons:
            output = neuron.feedforward(inputs)
            outputs.append(output)

        return outputs
    

class NeuralNetwork:                # Neural network has many layers
    def __init__(self, layers):
        self.layers = layers
    
    def feedforward(self, inputs):  # Outputs of one layer become inputs to the next layer
        outputs = inputs

        for layer in self.layers:
            outputs = layer.feedforward(outputs)
        
        return outputs
```

```python
import random

def build_neural_network(layer_sizes):
    layers = []
    for l in range(1, len(layer_sizes)):
        neurons = []
        num_inputs = layer_sizes[l - 1]
        num_neurons = layer_sizes[l]
        for _ in range(num_neurons):
            weights = [random.random() for _ in range(num_inputs)]
            bias = random.random()
            neuron = Neuron(weights, bias)
            neurons.append(neuron)
        
        layer = Layer(neurons)
        layers.append(layer)
    
    return NeuralNetwork(layers)

neural_network = build_neural_network(layer_sizes)

# Hidden neuron 1 -> NOT(x1 AND x2)
neural_network.layers[0].neurons[0].weights = [-1, -1]
neural_network.layers[0].neurons[0].bias = 1.5

# Hidden neuron 2 -> x1 OR x2
neural_network.layers[0].neurons[1].weights = [1, 1]
neural_network.layers[0].neurons[1].bias = -0.5

# Output neuron -> (NOT(x1 AND x2)) AND (x1 OR x2)
neural_network.layers[1].neurons[0].weights = [1, 1]
neural_network.layers[1].neurons[0].bias = -1.5

tests = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

for test in tests:
    output = neural_network.feedforward(test)
    print(test, "->", round(output[0], 5))
```
